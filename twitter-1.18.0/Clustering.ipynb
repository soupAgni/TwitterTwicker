```python
import re
#print jsons['tweetno']['text'][0]
allTweets = []

for tweetNum in jsons['tweetno']:
    result = re.sub(r"http\S+", "", tweetNum['text'])
    allTweets.append(result) 

print allTweets
```

Compute TF-IDF scores for each word in each item in array allTweets
